### 1.1. Создание графа, добавление узлов и рёбер, визуализация

## 2. Тренажёр работы с графами

```python
# Импорт библиотек.
import matplotlib.pyplot as plt
import networkx as nx
import random as rnd

# 0. Построение топологии по умолчанию.
def buildDefaultTopology():
    try:
        # Вывод списка топологий.
        print()
        print("Список топологий: ")
        print("1 - Полный граф Kn.")
        print("2 - Линейный граф Pn.")
        print("3 - Циклический граф Cn.")
        print("4 - Граф-колесо Wn.")
        print("5 - Звездообразный граф с n узлами.")
        print("6 - Полное бинарное дерево с n узлами.")
        print()
        # Ввод номера топологии с валидацией.
        topology = int(input("Выберите топологию: "))
        if (topology < 1 or topology > 6):
            print("Некорректный номер топологии!")
            return
        # Ввод числа узлов с валидацией.
        n = int(input("Выберите число узлов n: "))
        if (n < 0):
            print("Некорректное число узлов!")
            return
        global G
        # Построение указанной топологии.
        if (topology == 1):
            G = nx.complete_graph(n)
        elif (topology == 2):
            G = nx.path_graph(n)
        elif (topology == 3):
            G = nx.cycle_graph(n)
        elif (topology == 4):
            G = nx.wheel_graph(n)
        elif (topology == 5):
            G = nx.star_graph(n)
        else:
            G = nx.full_rary_tree(2, n)
        # Генерация весов рёбер графа.
        for (u, v) in G.edges():
            G.edges[u,v]['weight'] = rnd.randint(0, 100)
        print("Топология по умолчанию построена.")
    except:
        print("Некорректный ввод!")

# 1. Добавление одного узла.    
def addOneNode():
    # Ввод узла с валидацией.
    try:
        print()
        node = int(input("Введите номер узла: "))
        # Проверка наличия узла в графе.
        if (G.has_node(node)):
            print("Граф уже содержит указанный узел!")
        else:
            # Добавление узла.
            G.add_node(node)
            print("Узел добавлен в граф.")
    except:
        print("Некорректное значение узла!")

# 2. Добавление нескольких узлов.
def addNodes():
    # Ввод массива узлов с валидацией.
    try:
        print()
        nodes = list(map(int, input("Введите номера узлов через пробел: ").split()))
        contains = False
        # Проверка наличия узлов в графе.
        for node in nodes:
            if (G.has_node(node)):
                print("Граф уже содержит один из указанных узлов!")
                contains = True
                break
        if (not contains):
            # Добавление узлов.
            G.add_nodes_from(nodes)
            print("Узлы добавлены в граф.")
    except:
        print("Некорректные значения узлов!")

# 3. Добавление одного ребра.
def addOneEdge():
    # Ввод ребра с валидацией.
    try:
        print()
        node1 = int(input("Введите номер исходного узла: "))
        node2 = int(input("Введите номер конечного узла: "))
        # Проверка наличия петли.
        if (node1 == node2):
            print("Исходный и конечный узел совпадают!")
        # Проверка наличия ребра в графе.
        elif (G.has_edge(node1, node2)):
            print("Граф уже содержит указанное ребро!")
        else:
            inputWeight = int(input("Введите вес ребра: ")) 
            # Добавление ребра.
            G.add_edge(node1, node2, weight = inputWeight)
            print("Ребро добавлено в граф.")
    except:
        print("Некорректные значения ребра!")
          
# 4. Добавление нескольких рёбер.
def addEdges():
    # Ввод рёбер с валидацией.
    try:
        print()
        edges = list(tuple(map(int, input("Введите ребро в формате 'node1 node2 weight': ").split()))
                     for e in range(int(input('Введите число добавляемых рёбер: '))))
        error = False
        # Проверка корректности ввода и наличия рёбер в графе.
        for edge in edges:
            if (len(edge) != 3):
                raise Exception()
            if (edge[0] == edge[1]):
                print("Исходный и конечный узел совпадают!")
                error = True
                break
            if (G.has_edge(edge[0], edge[1])):
                print("Граф уже содержит одно из указанных рёбер!")
                error = True
                break
        if (not error):
            # Добавление рёбер.
            G.add_weighted_edges_from(edges)
            print("Рёбра добавлены в граф.")
    except:
        print("Некорректные значения рёбер!")
      
# 5. Удаление одного узла.
def removeOneNode():
    # Ввод узла с валидацией.
    try:
        print()
        node = int(input("Введите номер узла: "))
        # Проверка наличия узла в графе.
        if (not G.has_node(node)):
            print("Граф не содержит указанный узел!")
        else:
            # Удаление узла.
            G.remove_node(node)
            print("Узел удалён из графа.")
    except:
        print("Некорректное значение узла!")
   
# 6. Удаление нескольких узлов.
def removeNodes():
    # Ввод узлов с валидацией.
    try:
        print()
        nodes = list(map(int, input("Введите номера узлов через пробел: ").split()))
        contains = True
        for node in nodes:
            # Проверка наличия узлов в графе.
            if (not G.has_node(node)):
                print("Граф не содержит один из указанных узлов!")
                contains = False
                break
        if (contains):
            # Удаление узлов.
            G.remove_nodes_from(nodes)
            print("Узлы удалены из графа.")
    except:
        print("Некорректные значения узлов!")

# 7. Удаление одного ребра.
def removeOneEdge():
    # Ввод ребра с валидацией.
    try:
        print()
        node1 = int(input("Введите номер исходного узла: "))
        node2 = int(input("Введите номер конечного узла: "))
        # Проверка наличия ребра в графе.
        if (not G.has_edge(node1, node2)):
            print("Граф не содержит указанное ребро!")
        else:
            # Удаление ребра.
            G.remove_edge(node1, node2)
            print("Ребро удалено из графа.")
    except:
        print("Некорректные значения ребра!")
  
# 8. Удаление нескольких рёбер.
def removeEdges():
    # Ввод рёбер с валидацией.
    try:
        print()
        edges = list(tuple(map(int, input("Введите ребро в формате 'node1 node2': ").split()))
                     for e in range(int(input('Введите число удаляемых рёбер: '))))
        contains = True
        # Проверка корректности ввода и наличия рёбер в графе.
        for edge in edges:
            if (len(edge) != 2):
                raise Exception()
            if (not G.has_edge(edge[0], edge[1])):
                print("Граф не содержит одно из указанных рёбер!")
                contains = False
                break
        if (contains):
            # Удаление рёбер.
            G.remove_edges_from(edges)
            print("Рёбра удалены из графа.")
    except:
        print("Некорректные значения рёбер!")
   
# 9. Сохранение топологии в файл.
def saveToFile():
    # Запись в файл.
    nx.write_gml(G, "graph.gml")
    print()
    print("Топология записана в файл graph.gml.")

# 10. Считывание топологии из файла.
def readFromFile():
    # Чтение из файла с проверкой наличия.
    try:
        print()
        global G
        G = nx.read_gml("graph.gml", destringizer = int)
        print("Топология загружена из файла graph.gml")
    except:
        print("Файл graph.gml не найден!")

# 11. Вывод и сохранение топологии в графическом виде.
def visualizeGraph():
    # Заголовок.
    plt.title("Результат визуализации графа G:")
    # Граф.
    pos = nx.circular_layout(G)
    nx.draw(G,
            pos,
            node_color = 'skyblue',
            node_size = 500,
            with_labels = True)
    # Веса рёбер.
    edge_weight = nx.get_edge_attributes(G, 'weight')
    nx.draw_networkx_edge_labels(G, pos, edge_labels = edge_weight)
    # Сохранение топологии в графический файл.
    plt.savefig("graph.png")
    # Вывод топологии на экран.
    plt.show()
    
# 12. Вывод информации об узлах и рёбрах.
def printNodesAndEdges():
    # Узлы графа.
    print('\nУзлы графа G =', G.nodes())
    # Рёбра графа.
    print('Рёбра графа G =', G.edges.data("weight"))
   
# 13. Вывод информации о соседях узла.
def printNeighbors():  
    # Ввод номера узла с валидацией.
    try:
        print()
        node = int(input("Введите номер узла: "))
        # Проверка наличия узла в графе.
        if (G.has_node(node)):
            # Вывод соседних узлов.
            print('Соседи узла {}: {}'.format(node, G[node]))
        else:
            print("Граф не содержит указанный узел!")
    except:
        print("Некорректное значение узла!")

# 14. Выход из программы.
def closeProgram():
    print("\nЗавершение работы программы...")

# Создание пустого графа.
G = nx.Graph()

# Вывод списка операций.
print("Список операций: ")
print("0 - Построение топологии по умолчанию.")
print("1 - Добавление одного узла.")
print("2 - Добавление нескольких узлов.")
print("3 - Добавление одного ребра.")
print("4 - Добавление нескольких рёбер.")
print("5 - Удаление одного узла.")
print("6 - Удаление нескольких узлов.")
print("7 - Удаление одного ребра.")
print("8 - Удаление нескольких рёбер.")
print("9 - Сохранение топологии в файл.")
print("10 - Считывание топологии из файла.")
print("11 - Вывод и сохранение топологии в графическом виде.")
print("12 - Вывод информации об узлах и рёбрах.")
print("13 - Вывод информации о соседях узла.")
print("14 - Выход из программы.")

# Выполнение программы.
while (True):
    operation = -1
    # Ввод номера операции с валидацией.
    while (operation < 0 or operation > 14):
        try:
            print()
            operation = int(input("Выберите операцию: "))
        except:
            operation = -1
    
    # Выполнение указанной операции.
    if (operation == 0):
        buildDefaultTopology()
    if (operation == 1):
        addOneNode()
    elif (operation == 2):
        addNodes()
    elif (operation == 3):
        addOneEdge()
    elif (operation == 4):
        addEdges()
    elif (operation == 5):
        removeOneNode()
    elif (operation == 6):
        removeNodes()
    elif (operation == 7):
        removeOneEdge()
    elif (operation == 8):
        removeEdges()
    elif (operation == 9):
        saveToFile()
    elif (operation == 10):
        readFromFile()
    elif (operation == 11):
        visualizeGraph()
    elif (operation == 12):
        printNodesAndEdges()
    elif (operation == 13):
        printNeighbors()
    elif (operation == 14):
        closeProgram()
        break
```
