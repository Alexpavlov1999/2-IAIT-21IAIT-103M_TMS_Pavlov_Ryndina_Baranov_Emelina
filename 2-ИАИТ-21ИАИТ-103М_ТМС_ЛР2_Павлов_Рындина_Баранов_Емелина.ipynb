### 7. Доступ к рёбрам и соседям

```python
#Импорт
import matplotlib.pyplot as plt
import networkx as nx
#Создание графа
G = nx.Graph([(1, 2, {"color": "yellow"})])
#Данные о графе
print('G =',G)
#Данные о смежных узлах
print('G[1] =', G[1])
#Данные о конкретном смежном узле
print('G[1][2] =', G[1][2])
print('G.edges[1, 2] =', G.edges[1, 2])
```

```python
#Добавление ребра и задание цвета узлов
G.add_edge(1, 5)
G[1][5]['color'] = "green"
G.edges[1, 2]['color'] = "skyblue"
G.edges[1, 2]
```

```python
#Создание взвешенного графа
FG = nx.Graph()
FG.add_weighted_edges_from([(1, 2, 0.8), (1, 3, 2.7), (2, 4, 1.5), (3, 4, 0.6)])
#Вывод пар узлов с весами меньше 1
for n, nbrs in FG.adj.items():
   for nbr, eattr in nbrs.items():
       wt = eattr['weight']
       if wt < 1: print(f"({n}, {nbr}, {wt:.3})")
```

```python
#Вывод рёбер с весами меньше 1
for (u, v, wt) in FG.edges.data('weight'):
    if wt < 1:
        print(f"({u}, {v}, {wt:.3})")
```

### 11. Мультиграфы

```python
#Создание мультиграфа
MG = nx.MultiGraph()
MG.add_weighted_edges_from([(1, 2, 1.5), (1, 3, 2.5), (2, 3, 0.65)])
dict(MG.degree(weight = 'weight'))
GG = nx.Graph()
for n, nbrs in MG.adjacency():
   for nbr, edict in nbrs.items():
       minvalue = min([d['weight'] for d in edict.values()])
       GG.add_edge(n, nbr, weight = minvalue)
#Поиск кратчайшего пути из 1 в 3
nx.shortest_path(GG, 1, 3)
```

```python
#Импорт
import matplotlib.pyplot as plt
#Визуалиация
plt.title("MG.add_weighted_edges_from([(1, 2, 1.5), (1, 3, 2.5), (2, 3, 0.65)])")
#Список всех весов ребер
weights = nx.get_edge_attributes(MG, 'weight')
#Позиции вершин для визуализации
pos = nx.circular_layout(MG)
#Вывод графа
nx.draw_networkx(MG, pos = pos)
nx.draw_circular(DG,
         node_color = 'skyblue',
         node_size = 1000,
         with_labels = True)
plt.show()
```
