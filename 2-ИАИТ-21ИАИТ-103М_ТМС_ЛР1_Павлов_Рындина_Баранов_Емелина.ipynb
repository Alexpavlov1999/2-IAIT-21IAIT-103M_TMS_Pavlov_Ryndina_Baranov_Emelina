### 1. Создание графа с добавлением узлов и рёбер (по одному)

#### 1.1. Создание пустого графа

```python
#Импорт
import matplotlib.pyplot as plt
import networkx as nx
#Создание графа
G = nx.Graph()
#Вывод узлов и рёбер
print('Узлы сети =', G.nodes(), ';', 'Количество узлов сети =', len(G.nodes()))
print('Рёбра сети =', G.edges(), ';', 'Количество рёбер сети =', len(G.edges()))
```

#### 1.2. Добавление узлов (по одному)

```python
#Добавление узлов (метод add_node())
G.add_node('1')
G.add_node('2')
G.add_node('3')
G.add_node('4')
#Вывод узлов и рёбер
print('Узлы сети =', G.nodes(), ';', 'Количество узлов сети =', len(G.nodes()))
print('Рёбра сети =', G.edges(), ';', 'Количество рёбер сети =', len(G.edges()))
```

#### 1.3. Визуализация графа

```python
#Заголовок
plt.title("Узлы сети = ['1', '2', '3', '4']\nРёбра отсутствуют")
#Расположение узлов по кругу (метод nx.draw_circular())
nx.draw_circular(G,
         node_color = 'skyblue',
         node_size = 1000,
         with_labels = True)
#Вывод графа
plt.show()
```

#### 1.4. Добавление рёбер (по одному)

```python
#Добавление рёбер (метод add_edge())
G.add_edge('1', '2')
G.add_edge('1', '3')
G.add_edge('2', '3')
G.add_edge('2', '4')
#Вывод узлов и рёбер
print('Узлы сети =', G.nodes(), ';', 'Количество узлов сети =', len(G.nodes()))
print('Рёбра сети =', G.edges(), ';', 'Количество рёбер сети =', len(G.edges()))
#Заголовок
plt.title("Результат добавления рёбер (по одному)")
#Расположение узлов по кругу (метод nx.draw_circular())
nx.draw_circular(G,
         node_color = 'skyblue',
         node_size = 1000,
         with_labels = True)
#Вывод графа
plt.show()
```

#### 1.5. Вывод количества и имён соседей

```python
#Вывод количества и имён соседей
print('Соседи узла 1 =', list(G.neighbors('1')))
print('Соседи узла 2 =', list(G.neighbors('2')))
print('Соседи узла 3 =', list(G.neighbors('3')))
print('Соседи узла 4 =', list(G.neighbors('4')))
print('Число соседей узла 1 =', G.degree('1'))
print('Число соседей узла 2 =', G.degree('2'))
print('Число соседей узла 3  =', G.degree('3'))
print('Число соседей узла 4  =', G.degree('4'))
```

### 2. Создание графа с добавлением узлов и рёбер (по несколько)

#### 2.1. Создание пустого графа, добавление узлов (по несколько)

```python
#Импорт
import matplotlib.pyplot as plt
import networkx as nx
#Создание графа
G = nx.Graph()
#Добавление нескольких узлов 
G.add_nodes_from([1, 2, 3, 4, 5, 6, 7, 8])
#Вывод узлов
print('Узлы сети = ', G.nodes())
#Вывод графа
nx.draw_circular(G,
         node_color = 'skyblue',
         node_size = 1000,
         with_labels = True)
plt.show()
```

#### 2.2. Добавление рёбер (по несколько)

```python
#Добавление нескольких рёбер
G.add_edges_from([(1, 2), (1, 3), (1, 6), (2, 4), (2, 6), (2, 8), (3, 7), (4, 5), (5, 7), (7, 8)])
#Вывод рёбер
print('Рёбра сети', G.edges())
#Вывод графа
nx.draw_circular(G,
         node_color = 'skyblue',
         node_size = 1000,
         with_labels = True)
plt.show()
```

### 3. Удаление узлов из графа

#### 3.1. Создание графа, удаление одного узла

```python
#Импорт
import matplotlib.pyplot as plt
import networkx as nx
#Создание графа
G = nx.Graph()
#Добавление узлов
G.add_nodes_from([1, 2, 3, 4, 5, 6, 7, 8])
#Добавление рёбер
G.add_edges_from([(1, 2), (1, 3), (1, 6), (2, 4), (2, 6), (2, 8), (3, 7), (4, 5), (5, 7), (7, 8)])
#Вывод узлов и рёбер
print('До удаления узла 3:')
print('Узлы сети =', G.nodes())
print('Рёбра сети =', G.edges())
#Удаление узла 3
G.remove_node(3)
#Вывод узлов и рёбер
print('После удаления узла 3:')
print('Узлы сети =', G.nodes())
print('Рёбра сети =', G.edges())
#Вывод графа
nx.draw_circular(G,
         node_color = 'skyblue',
         node_size = 1000,
         with_labels = True)
plt.show()
```

#### 3.2. Удаление нескольких узлов

```python
#Удаление узлов 6, 7, 8
G.remove_nodes_from([6, 7, 8])
#Вывод узлов и рёбер
print('После удаления узлов 6, 7, 8:')
print('Узлы сети =', G.nodes())
print('Рёбра сети =', G.edges())
#Вывод графа
nx.draw_circular(G,
         node_color = 'skyblue',
         node_size = 1000,
         with_labels = True)
plt.show()
```

### 4. Удаление рёбер из графа

#### 4.1. Создание графа

```python
#Импорт
import matplotlib.pyplot as plt
import networkx as nx
#Создание графа
G = nx.Graph()
#Добавление узлов и рёбер
G.add_nodes_from([1, 2, 3, 4, 5, 6, 7, 8])
G.add_edges_from([(1, 2), (1, 3), (1, 6), (2, 4), (2, 6), (2, 8), (3, 7), (4, 5), (5, 7), (7, 8)])
#Вывод графа
plt.title("Исходный граф")
nx.draw_circular(G,
         node_color = 'skyblue',
         node_size = 1000,
         with_labels = True)
plt.show()
```

#### 4.2. Удаление первого ребра

```python
#Удаление одного ребра
G.remove_edge(3, 7)
#Вывод графа
plt.title("Результат удаления ребра (3, 7)")
nx.draw_circular(G,
         node_color = 'skyblue',
         node_size = 1000,
         with_labels = True)
plt.show()
```

#### 4.3. Удаление второго ребра

```python
#Удаление одного ребра
G.remove_edge(1, 6)
#Вывод графа
plt.title("Результат удаления ребра (1, 6)")
nx.draw_circular(G,
         node_color = 'skyblue',
         node_size = 1000,
         with_labels = True)
plt.show()
```

#### 4.4. Удаление нескольких рёбер

```python
#Удаление нескольких рёбер
G.remove_edges_from([(1, 2), (2, 8), (4, 5)])
#Вывод графа
plt.title("Результат удаления рёбер (1, 2), (2, 8), (4, 5)")
nx.draw_circular(G,
         node_color = 'skyblue',
         node_size = 1000,
         with_labels = True)
plt.show()
```
